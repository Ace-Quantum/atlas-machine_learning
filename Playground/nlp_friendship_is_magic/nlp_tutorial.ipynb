```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer

X = ["This is a sentance",
     "And Another one",
     "How is your day?",
     "Are you looking forward to pizza later?",
     "Have you gotten enough sleep?",
     "No? Well that's a problem",
     "Oh hey, a fun dataset for this would be character dialogue for videogames",
     "That's an actual job I could get in ML, huh?"]

vectorizer = CountVectorizer()
vectorizer.fit(X)
```

```python
vectorizer.vocabulary_
# Hey um
# This isn't sorted in any way, is it?
# That's fun.
```

```python
bag_o_words = vectorizer.transform(X)
```

```python
bag_o_words.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid_vec = TfidfVectorizer()
tfid_vec.fit(X)
```

```python
import numpy as np
np.set_printoptions(precision=2)

print(tfid_vec.transform(X).toarray())
```
